```python
p2pc_dummies = pd.get_dummies(df_model_type.P2PC)
p2pb_dummies = pd.get_dummies(df_model_type.P2PB)
p2pp_dummies = pd.get_dummies(df_model_type.P2PP)
df_model_frame = pd.concat([df_model_type.ResponseID, p2pc_dummies, p2pb_dummies, p2pp_dummies], axis =1)
```

```python
df_model_frame
```

```python
df = df_model_frame.groupby(['ResponseID']).count()
```

```python
df.plot()
plt.xticks(rotation=45) 
plt.show() 
```
